## 1. Project objective

The goal of this project was to forecast the NZD/USD exchange rate using historical daily exchange-rate data.

The project followed a realistic forecasting workflow:

1. Clean and transform the raw time-series data.
2. Build simple baseline forecasts.
3. Analyze stationarity and autocorrelation.
4. Fit ARIMA-style models.
5. Convert the time series into a supervised learning problem.
6. Compare machine-learning models against strong baselines.

## 2. Key modeling results

| Modeling stage | Best-performing approach | Main finding |
|---|---|---|
| Baseline forecasting | Naïve / drift baselines | Simple baselines were difficult to beat. |
| ARIMA modeling | Drift baseline | ARIMA models did not outperform the drift baseline in the fixed-origin test setup. |
| Machine-learning forecasting | Drift baseline | Linear, ridge, and random forest models did not beat one-step-ahead baselines. |

## 3. Important evaluation note

The ARIMA and machine-learning notebooks used different forecasting setups.

The ARIMA notebook used a fixed-origin multi-step forecast, where models were trained up to the training cutoff and then forecast across the full test period.

The machine-learning notebook used a one-step-ahead supervised learning setup, where each row used current and past information to predict the next observed exchange rate.

Because of this difference, ARIMA and machine-learning metrics should not be compared directly without considering the forecasting setup. Within each setup, models were compared against appropriate baselines.

## 4. Main conclusion

Across the project, simple baseline models remained very difficult to beat.

The exchange-rate level was highly persistent, meaning the current exchange rate was usually the strongest predictor of the next observed exchange rate. Stationarity and autocorrelation analysis showed that daily returns had weak linear dependence, which helped explain why more complex models struggled to add predictive value.

The ARIMA models produced only small improvements over a random-walk specification and did not beat the drift baseline in the fixed-origin test setup. The supervised machine-learning models also failed to outperform one-step-ahead naïve and drift baselines.

Overall, the project shows that more complex forecasting models do not automatically produce better forecasts. Strong baseline comparisons are essential, especially for financial time series such as exchange rates.
